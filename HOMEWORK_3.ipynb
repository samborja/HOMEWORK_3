```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tesla = pd.read_csv("C:/Users/JAIRO BORJA/Desktop/UNIVERSIDAD/2SEMESTRE/INTELIGENCIA/TSLA.csv")
tesla.head(1)
```

```python
tesla[['Open','High','Low','Close','Adj Close','Volume']].describe()
```

```python
#Escalado
tesla['OpenA'] = tesla['Open'] / tesla['Open'].max()
tesla['HighA'] = tesla['High'] / tesla['High'].max()
tesla['LowA'] = tesla['Low'] / tesla['Low'].max()
tesla['CloseA'] = tesla['Close'] / tesla['Close'].max()

#Min-Max
tesla['OpenB'] = (tesla['Open'] - tesla['Open'].min())/(tesla['Open'].max() - tesla['Open'].min())
tesla['HighB'] = (tesla['High'] - tesla['High'].min())/(tesla['High'].max() - tesla['High'].min())
tesla['LowB'] = (tesla['Low'] - tesla['Low'].min())/(tesla['Low'].max() - tesla['Low'].min())
tesla['CloseB'] = (tesla['Close'] - tesla['Close'].min())/(tesla['Close'].max() - tesla['Close'].min())

#Z-score
tesla['OpenC'] = (tesla['Open'] - tesla['Open'].mean()) / tesla['Open'].std()
tesla['HighC'] = (tesla['High'] - tesla['High'].mean()) / tesla['High'].std()
tesla['LowC'] = (tesla['Low'] - tesla['Low'].mean()) / tesla['Low'].std()
tesla['CloseC'] = (tesla['Close'] - tesla['Close'].mean()) / tesla['Close'].std()

tesla[['OpenA', 'OpenB', 'OpenC']]
```
